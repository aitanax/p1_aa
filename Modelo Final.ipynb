```python
import pandas as pd
from joblib import load
```

```python
modelo_final = load('modelo_final.pkl')
```

Leemos los datos de la competición:

```python
pd.set_option('display.max_columns', 6000)
pd.set_option('display.max_rows', 1000)

wind_comp = pd.read_csv('wind_comp.csv.gz', compression='gzip')
wind_comp.head()
```

```python
lista_col = ['datetime']
lista_col.extend(col for col in wind_comp.columns if col.endswith('.13'))
```

```python
wind_comp_13 = wind_comp[lista_col]
wind_comp_13.head()
```

```python
X = wind_comp_13.iloc[:, 1:]
pred_comp = modelo_final.predict(X)
pred_comp
```

```python
df = pd.DataFrame(pred_comp, columns=['predicciones'])
df.to_csv('predicciones.csv', index=False)
```
